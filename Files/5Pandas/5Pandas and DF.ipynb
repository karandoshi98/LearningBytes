```python
l = pd.DataFrame({'key':['k0','k8','k2','k3'],
                 'a':['a0','a1','a2','a3'],
                 'b':['b0','b1','b2','b3']})

r = pd.DataFrame({'key':['k0','k1','k2','k3'],
                 'c':['c0','c1','c2','c3'],
                 'd':['d0','d1','d2','d3']})
```

```python
pd.merge(l,r)
```

```python
pd.merge(l,r,on='key') #default - inner join merge
```

```python
pd.merge(l,r,on='key', how = 'left') # left outer join
```

```python
pd.merge(l,r,on='key', how = 'outer') # full outer join
```

### Join

```python
l.join(r,on='key')
```

```python
# We either need to have different names for columns or specify suffixes (l and r)
```

```python
l = pd.DataFrame({'key':['k0','k8','k2','k3'],
                 'a':['a0','a1','a2','a3'],
                 'b':['b0','b1','b2','b3']},
                    index=['a0','a1','a2','a3'])

r = pd.DataFrame({'key':['k0','k1','k2','k3'],
                 'c':['c0','c1','c2','c3'],
                 'd':['d0','d1','d2','d3']},
                    index=['a0','a1','a2','a3'])

l.join(r,lsuffix=1,rsuffix=2)
```

```python
l = pd.DataFrame({'key':['k0','k8','k2','k3'],
                 'a':['a0','a1','a2','a3'],
                 'b':['b0','b1','b2','b3']},
                    index=['a0','a1','a2','a3'])

r = pd.DataFrame({'key122':['k0','k1','k2','k3'],
                 'c':['c0','c1','c2','c3'],
                 'd':['d0','d1','d2','d3']},
                    index=['a0','a1','a2','a3'])

l.join(r)
```

### Apply function on dataframe

```python
titanic.dropna(axis=1)
```

```python
titanic['Len_name'] = titanic['Name'].apply(len)
titanic
```

```python
#titanic['Sqr_Fare'] = titanic['Fare'].apply(np.square)
titanic['Sqr_Fare'] = titanic['Fare'].apply(lambda x:np.square(x))

titanic
```

```python
def test(x):
    return x+2

titanic['taxed_Fare'] = titanic['Fare'].apply(test)
titanic
```

```python
def categorical(x):
    if x == 'female':
        return 0
    else:
        return 1

titanic['categorical_sex'] = titanic['Sex'].apply(categorical)
titanic
```
