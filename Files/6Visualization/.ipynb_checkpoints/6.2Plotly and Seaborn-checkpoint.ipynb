# Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

##### Scatter plot

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue ='day',style='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,style='size')
```

##### Categorical vs Numerical plot

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='swarm')
```

```python
sns.catplot(x='day',y='tip',data=tips, kind='boxen')
```

```python
sns.catplot(x='day',y='tip',data=tips, kind='box',hue='sex')
```

```python
x = np.random.randn(1000)
x
```

##### Distribution of a plot

```python
sns.distplot(x)
```

```python
tips
```

##### JointPlot   -   Scatter plot + Histogram

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'], kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kdeplot')
```

##### Pairplot

```python
sns.pairplot(tips)
```

```python
tips
```

##### Lmplot - Plot data and regression model fits across a FacetGrid.

```python
sns.lmplot(x='total_bill',y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips , col ='day') 
# different graphs according to a categorical variable
```

##### Regplot - Linear regression plot across data

```python
sns.regplot(x='total_bill',y='tip', data=tips) 
```
